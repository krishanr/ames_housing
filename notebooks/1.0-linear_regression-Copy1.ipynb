## Rescale to unit variance

```python
# First take the logarithm of skewed distributions
columns_to_log = ["LotFrontage", "LotArea", "fossil_fuel_capacity", "annual_peak_load", "num_residential_accounts", "elev", "elec_rate_general", "estimated_min_genset"]
for column in columns_to_log:
    df[column + "_log"] = np.log1p(df[column])
    df[column + "_log"] = df[column + "_log"]/df[column + "_log"].std()
```

```python
df["latitude"] = df["latitude"]/df["latitude"].std()
df["longitude"] = df["longitude"]/df["longitude"].std()
df["elec_rate_household"] = df["elec_rate_household"]/df["elec_rate_household"].std()
df["cwec_station_dist"] = df["cwec_station_dist"]/df["cwec_station_dist"].std()
```
